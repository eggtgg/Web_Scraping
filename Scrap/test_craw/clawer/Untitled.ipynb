```python
import requests
from bs4 import BeautifulSoup
import codecs
```

```python
def reqweb(link,i):
    r=requests.get(link)
    soup= BeautifulSoup(r.text, 'html.parser')
    links=soup.find_all('div',attrs={'class':"d-ib w-400"})
    for link in links:
        file_html= codecs.open('file'+str(i)+'.html','w','utf8')
        file_html.write(requests.get('https://vietnamnet.vn'+link.find('a')['href']).text)
        file_html.close()
        i+=1
    c='https://vietnamnet.vn'+link.find('a')['href']
    reqweb(c,i)
reqweb('https://vietnamnet.vn/vn/kinh-doanh/tai-chinh/donald-trump-that-the-nguoi-dan-ong-quyen-luc-lo-dien-698112.html',0)
    
    
```
