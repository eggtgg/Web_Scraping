```
import requests
from bs4 import BeautifulSoup
```

```
#1 REQUESTS LINK
d= requests.get('https://vietnamnet.vn/vn/suc-khoe/do-nguy-hiem-cua-bien-chung-ncov-moi-tai-viet-nam-702459.html')
d.text
```

```
# GHI FILE

file = open('file' +'1'+ '.html','w')
file.write(d.text)
file.close()
```

```
# THỬ VỚI QUI MÔ NHỎ
def add_link(d,i):
    l = requests.get(d)

    link_html = BeautifulSoup(l.text, 'html.parser')
    a = link_html('div', class_="clearfix item")

    for a_i in a:
        linkrq ='https://vietnamnet.vn'+ a_i.find('a')['href']
        if linkrq in s:
            continue
        else:
            s.append(linkrq)
            file = open('file' + str(i)+ '.html','w')
            file.write(requests.get(linkrq).text)
            file.close()
            i += 1
            
```

```
s = ['https://vietnamnet.vn/vn/thoi-su/an-toan-giao-thong/xe-limousine-gay-tai-nan-tham-khoc-do-phong-nhanh-vuot-au-699177.html']
for s_i in s:
    add_link(s_i,0)
    
```
